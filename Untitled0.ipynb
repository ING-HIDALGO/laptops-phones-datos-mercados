Análisis de Ingresos por Marca, El objetivo es identificar las marcas líderes en ventas monetarias para priorizar alianzas comerciales.
¿Cuáles son las 5 marcas que generan más ingresos totales?

---



```python
# Consulta SQL usando la tabla ya limpia
query_1 = """
SELECT
    marca,
    SUM(precio * cantidad_vendida) AS ingresos_totales
FROM `analisis-de-datos-portafolio.analisis_de_datos_portafolio.telefonos_limpios`
GROUP BY marca
ORDER BY ingresos_totales DESC
LIMIT 5
"""


df_ingresos = client.query(query_1).to_dataframe()

print(df_ingresos)


df_ingresos.plot(kind='bar', x='marca', y='ingresos_totales', title='Top 5 Marcas por Ingresos')
```

¿Cuál es el tiempo promedio de despacho (en días) por marca?

---



```python
query_2 = """
SELECT
    marca,
    ROUND(AVG(DATE_DIFF(fecha_despacho, fecha_entrada, DAY)), 1) AS promedio_dias_entrega
FROM `analisis-de-datos-portafolio.analisis_de_datos_portafolio.telefonos_limpios`
GROUP BY marca
ORDER BY promedio_dias_entrega ASC
"""

df_logistica = client.query(query_2).to_dataframe()
print(df_logistica)

# Gráfico para comparar eficiencia
df_logistica.plot(kind='barh', x='marca', y='promedio_dias_entrega',
                  title='Días Promedio de Despacho por Marca (Menos es mejor)',
                  color='orange')
```

Podemos observar que el promedio de entrega entre las diferentes empresas es muy ajustado, esto deja ver la competencia en logistica de las marcas, los pequeños detalles hacen la diferencia, a quien pidio el estudio puedo decirle que tendriamos que hacer un analisis comparativo de estructura logistica solo contra la marca que lo adelanta en el ranking. revisando esto puedo darle pauta para adelantar la entrega.

¿Cómo se distribuyen las unidades vendidas por Región?

---



```python
query_3 = """
SELECT
    region,
    SUM(cantidad_vendida) AS unidades_totales
FROM `analisis-de-datos-portafolio.analisis_de_datos_portafolio.telefonos_limpios`
GROUP BY region
ORDER BY unidades_totales DESC
"""

df_region = client.query(query_3).to_dataframe()
print(df_region)

# Gráfico de pastel (Pie Chart) para ver la cuota de mercado por zona
df_region.set_index('region')['unidades_totales'].plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(8, 8),
    title='Distribución de Ventas por Región'
)
```

Consulta de Verificación (Pregunta 3.1):

Análisis de Inventario: ¿Cuál es el producto más vendido por cada marca?

---



```python
query_5 = """
WITH RankingVentas AS (
    SELECT
        marca,
        producto,
        SUM(cantidad_vendida) AS total_unidades,
        RANK() OVER(PARTITION BY marca ORDER BY SUM(cantidad_vendida) DESC) AS puesto
    FROM `analisis-de-datos-portafolio.analisis_de_datos_portafolio.telefonos_limpios`
    GROUP BY marca, producto
)
SELECT
    marca,
    producto,
    total_unidades
FROM RankingVentas
WHERE puesto = 1
ORDER BY total_unidades DESC
"""

df_top_productos = client.query(query_5).to_dataframe()
print(df_top_productos)

# Gráfico de barras comparativo
plt.figure(figsize=(10, 6))
plt.bar(df_top_productos['marca'] + " - " + df_top_productos['producto'],
        df_top_productos['total_unidades'], color='purple')
plt.xticks(rotation=45, ha='right')
plt.title('Producto Estrella por Marca (Basado en Unidades Vendidas)')
plt.ylabel('Unidades Vendidas')
plt.show()
```
